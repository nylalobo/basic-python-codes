```python
text = "Hello World"
letter_count = {}

for letter in text:
    if letter in letter_count:
        letter_count[letter] += 1
    else:
        letter_count[letter] = 1

print(letter_count)
```

```python
import cv2
import matplotlib.pyplot as plt

# Use the correct file name as a string
cap = cv2.VideoCapture("beach video.mp4")

if not cap.isOpened():
    print("Error: Could not open video.")
else:
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_num = int(4 * fps)
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_num)
    ret, frame = cap.read()

    if ret:
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        plt.imshow(frame_rgb)
        plt.title("Frame at 4 seconds")
        plt.axis("off")
        plt.show()
    else:
        print("Failed to read frame.")

    cap.release()
```

```python
import cv2
import imageio
from google.colab import files

# Correct video path (string)
video_path = "beach video.mp4"

# Open video file
cap = cv2.VideoCapture(video_path)

# Get frames per second
fps = cap.get(cv2.CAP_PROP_FPS)

# Total frames for 3 seconds
frame_count = int(fps * 3)

frames = []  # Initialize empty list to store frames

# Read frames
for i in range(frame_count):
    ret, frame = cap.read()
    if not ret:
        break
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frames.append(rgb)

cap.release()

# Save as GIF
gif_path = "clip.gif"
imageio.mimsave(gif_path, frames, fps=int(fps))

# Download the GIF
files.download(gif_path)
```

```python
import cv2

filename = "beach video.mp4"  # or whatever your video file is

cap = cv2.VideoCapture(filename)

if not cap.isOpened():
    print("Error: Could not open video.")
else:
    print("Width:", int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)))
    print("Height:", int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
    print("Frame Rate (FPS):", cap.get(cv2.CAP_PROP_FPS))
    print("Total Frames:", int(cap.get(cv2.CAP_PROP_FRAME_COUNT)))

    cap.release()
```
